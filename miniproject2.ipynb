```python
import statistics
```

```python
scores = [64, 56, 68, 71, 75, 78, 82, 85, 88, 90, 92]
study_hours = [2, 3, 3, 4, 4, 5, 6, 6, 7, 8, 8]
```

```python
mean = statistics.mean(scores)
print(mean)
```

```python
median = statistics.median(scores)
print(median)
```

```python
mode = statistics.mode(scores)
print(mode)
```

```python
variance = statistics.variance(scores)
print(variance)
```

```python
std = statistics.stdev(scores)
print(std)
```

```python
median_low = statistics.median_low(scores)
median_high = statistics.median_high(scores)
```

```python
paired_data = list(zip(study_hours, scores))
print(paired_data)
```

```python
statistics.fmean(scores)
```
